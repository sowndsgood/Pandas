```python
df=pd.read_csv("D:\\Pandas\\chapter_3_assets\\4_handling_missing_data_fillna_dropna_interpolate\\weather_data.csv",parse_dates=["day"])
df.head(3)
```

```python
type(df.day[0])
```

```python
df.set_index('day',inplace=True)   #inplace is important argument to modify the df originally
df
```

```python
df.fillna(0,inplace=True) #replace null with 0
df
```

```python
new_df = df.fillna({                                         #use mean or median
        'temperature': df.temperature.mean(),
        'windspeed': df.windspeed.mean(),
        'event': 'No Event'
    })
new_df
```

```python
df1=df.fillna(method='ffill')
df1
```

```python
df2=df.fillna(method='bfill')
df2
```

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
new_df
```

```python
new_df = df.fillna(method="ffill",limit=1)
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
new_df=df.dropna(how="any")  #drops if any is na
new_df
```

```python
new_df=df.dropna(how="all")     #drops only if alll is na
new_df 
```

```python
new_df=df.dropna(thresh=1)  #threshold
new_df
```
