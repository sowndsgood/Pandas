```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("D:\\Pandas\\chapter_3_assets\\5_handling_missing_data_replace\\weather_data.csv")
df
#-99999 is used as nulll value
```

```python
df_new =df.replace(-99999,0)
df_new
```

```python
df_new =df.replace([-99999,-88888],np.NaN)
df_new
```

```python
new_df = df.replace(to_replace=[-99999,-88888], value=0)
new_df
```

```python
new_df=df.replace({
    'temperature': -99999,
    'windspeed': [-99999,-88888],
    'event': 'no event'
},np.NaN)
new_df
```

```python
new_df = df.replace({
        -99999: np.nan,  #dict with old value replaced by new value
        'no event': 'Sunny',
    })
new_df
```
