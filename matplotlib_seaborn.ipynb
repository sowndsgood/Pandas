```python
import pandas as pd

df_sales = pd.read_excel("D:\\Pandas\\chapter_3_assets\\8_matplotlib_seaborn\\linechart.xlsx")
df_sales.head()
```

```python
from matplotlib import pyplot as plt
plt.plot(df_sales["Quarter"],df_sales["Fridge"])
```

```python
#To remove first line of op
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"])
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"])
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"])
plt.show()
```

```python
#To give specific color
plt.figure(figsize=(12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],color="green",label="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],color="red",label="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],color="yellow",label="Washing Machine")
plt.title("Product Sales")
plt.xlabel("Product")
plt.ylabel("Quarter")
plt.legend()
plt.show()
```

# Pie Chart

```python
df=df_sales[["Fridge","Dishwasher","Washing Machine"]].sum()
df
```

```python
plt.pie(df,labels=df.index,autopct='%1.1f%%',shadow=True,explode=(0.1,0,0),startangle=140)
plt.show()
```

# Bar Chart

```python
df_sales.plot(kind='bar',x="Quarter")
plt.xticks(rotation=0)
plt.show()
```

```python
#To use Quarters as X axis, we can set them as an index on a dataframe and then use plot function
df_sales_2 = df_sales.set_index("Quarter")
df_sales_2
```

```python
df_sales_2.plot(kind="bar") #index will be deault X axis
plt.xticks(rotation=45)
plt.show()
```

```python
df_sales_2.plot(kind="barh")
plt.show()
```
